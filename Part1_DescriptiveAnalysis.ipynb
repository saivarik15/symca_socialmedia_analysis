```python
import hashlib
import pandas as pd

# Define a function to generate a unique ID using SHA-256 hash
def generate_handlename(data):
    sha256_hash = hashlib.sha256()
    sha256_hash.update(data.encode('utf-8'))
    hex_digest = sha256_hash.hexdigest()
    return hex_digest

# Define a function to process a file and create the selected_columns_df
def process_file(file_path):
    # Read the data from the Parquet file
    df_tweets = pd.read_parquet(file_path)
    
    # Filter the data based on the 'Handle' column
    filtered_df = df_tweets[df_tweets['Handle'] != '@SouthYorksMCA']
    filtered_council_df = df_tweets[df_tweets['Handle'] == '@SouthYorksMCA']
    
    # Add a 'unique_id' column based on the 'Handle'
    filtered_df["unique_id"] = filtered_df["Handle"].apply(generate_handlename)
    filtered_council_df["unique_id"] = '@SouthYorksMCA'
    
    # Concatenate the filtered DataFrames
    combined_df = pd.concat([filtered_council_df, filtered_df], ignore_index=True)
    
    # Select the desired columns
    selected_columns_df = combined_df[['Date_Time', 'Tweet', 'Reply', 'Retweet', 'Likes', 'unique_id']]
    # Assuming you have already created the selected_columns_df DataFrame
    selected_columns_df.rename(columns={'unique_id': 'Handle'}, inplace=True)

    
    return selected_columns_df

# File paths for the different datasets
file_path_SYMCA = './Dataset/SYMCA2.parquet'
file_path_Essex = './Dataset/Essex2.parquet'
file_path_Hampshire = './Dataset/Hamisphere2.parquet'
file_path_Surrey = './Dataset/Surrey2.parquet'

# Process each file and create DataFrames with the specified names
df_tweets_SYMCA = process_file(file_path_SYMCA)
df_tweets_Essex = process_file(file_path_Essex)
df_tweets_Hampshire = process_file(file_path_Hampshire)
df_tweets_Surrey = process_file(file_path_Surrey)

# Now you have four separate DataFrames with the specified names
```

```python
# File paths for writing the Parquet files
output_path_SYMCA = './Dataset/SYMCA.parquet'
output_path_Essex = './Dataset/Essex.parquet'
output_path_Hampshire = './Dataset/Hampshire.parquet'
output_path_Surrey = './Dataset/Surrey.parquet'

# Write the DataFrames to Parquet files with the specified names
df_tweets_SYMCA.to_parquet(output_path_SYMCA, index=False)
df_tweets_Essex.to_parquet(output_path_Essex, index=False)
df_tweets_Hampshire.to_parquet(output_path_Hampshire, index=False)
df_tweets_Surrey.to_parquet(output_path_Surrey, index=False)
```

```python
#Preprocessing

def map_to_numeric(likes_str):
    try:
        # Convert to string
        likes_str = str(likes_str)
        
        if likes_str.endswith('K'):
            # Remove 'K' character and convert to integer
            return int(float(likes_str[:-1]) * 1000)
        elif likes_str.endswith('M'):
            # Remove 'M' character and convert to integer
            return int(float(likes_str[:-1]) * 1000000)
        else:
            # Convert to integer if no 'K' or 'M' is present
            return int(float(likes_str))
    except ValueError:
        # Handle non-integer values gracefully (e.g., return 0)
        return 0


    
# Apply the mapping function to the 'Likes' column for df_tweets_SYMCA
df_tweets_SYMCA['Likes'] = df_tweets_SYMCA['Likes'].apply(map_to_numeric)

# Apply the mapping function to the 'Retweets' column for df_tweets_SYMCA
df_tweets_SYMCA['Retweet'] = df_tweets_SYMCA['Retweet'].apply(map_to_numeric)

# Apply the mapping function to the 'Reply' column for df_tweets_SYMCA
df_tweets_SYMCA['Reply'] = df_tweets_SYMCA['Reply'].apply(map_to_numeric)

# Repeat the same process for df_tweets_Essex, df_tweets_Hampshire, and df_tweets_Surrey
df_tweets_Essex['Likes'] = df_tweets_Essex['Likes'].apply(map_to_numeric)
df_tweets_Essex['Retweet'] = df_tweets_Essex['Retweet'].apply(map_to_numeric)
df_tweets_Essex['Reply'] = df_tweets_Essex['Reply'].apply(map_to_numeric)

df_tweets_Hampshire['Likes'] = df_tweets_Hampshire['Likes'].apply(map_to_numeric)
df_tweets_Hampshire['Retweet'] = df_tweets_Hampshire['Retweet'].apply(map_to_numeric)
df_tweets_Hampshire['Reply'] = df_tweets_Hampshire['Reply'].apply(map_to_numeric)

df_tweets_Surrey['Likes'] = df_tweets_Surrey['Likes'].apply(map_to_numeric)
df_tweets_Surrey['Retweet'] = df_tweets_Surrey['Retweet'].apply(map_to_numeric)
df_tweets_Surrey['Reply'] = df_tweets_Surrey['Reply'].apply(map_to_numeric)



```

```python
import matplotlib.pyplot as plt
import pandas as pd

def calculate_tweet_counts(dataframe, council_name):
    # Replace '@SouthYorksMCA' with 'SouthYorksMCA' in the 'Handle' column
    dataframe['Handle'] = dataframe['Handle'].replace('@'+council_name, council_name)

    # Filter out specific users and keep the specified council name
    filtered_df = dataframe[dataframe['Handle'] == council_name]

    # Count the tweets by the specified council name and others
    tweet_counts = filtered_df['Handle'].value_counts().to_dict()

    # Combine the specified council name and others into 'Public'
    tweet_counts['Public'] = dataframe.shape[0] - tweet_counts.get(council_name, 0)

    return tweet_counts

# Calculate tweet counts for different councils
SYMCA_counts = calculate_tweet_counts(df_tweets_SYMCA, 'SouthYorksMCA')
Essex_counts = calculate_tweet_counts(df_tweets_Essex, 'Essex_CC')
Hampshire_counts = calculate_tweet_counts(df_tweets_Hampshire, 'hantsconnect')
Surrey_counts = calculate_tweet_counts(df_tweets_Surrey, 'SurreyCouncil')

# Create subplots for four bar charts
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle("Tweet Counts by Councils")

# Plot SYMCA_counts
SYMCA_counts_series = pd.Series(SYMCA_counts)
SYMCA_counts_series.plot(kind='bar', ax=axes[0, 0], title="SYMCA Counts")

# Plot Essex_counts
Essex_counts_series = pd.Series(Essex_counts)
Essex_counts_series.plot(kind='bar', ax=axes[0, 1], title="Essex Counts")

# Plot Hampshire_counts
Hampshire_counts_series = pd.Series(Hampshire_counts)
Hampshire_counts_series.plot(kind='bar', ax=axes[1, 0], title="Hampshire Counts")

# Plot Surrey_counts
Surrey_counts_series = pd.Series(Surrey_counts)
Surrey_counts_series.plot(kind='bar', ax=axes[1, 1], title="Surrey Counts")

# Adjust subplot layout
plt.tight_layout()
plt.subplots_adjust(top=0.85)

# Show the combined plot
plt.show()

# Calculate the total counts for all councils
total_counts = {
    'SouthYorksMCA': sum(SYMCA_counts.values()),
    'Essex_CC': sum(Essex_counts.values()),
    'hantsconnect': sum(Hampshire_counts.values()),
    'SurreyCouncil': sum(Surrey_counts.values())
}

# Create a pie chart for the total counts
labels = total_counts.keys()
sizes = total_counts.values()
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # Explode the 1st slice (SouthYorksMCA)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Total Tweet Counts by Councils")
plt.show()





```
